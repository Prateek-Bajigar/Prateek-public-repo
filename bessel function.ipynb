### Q1.  Plot Bessel functions of first kind $(J_n(x))$ with order n=0,1,2,3,4 wrt x. Given
$$ J_n(x) = \sum_{m=0}^{\infty} \frac{(-1)^m \left(\frac{x}{2}\right)^{2m+n}}{m!(m+n)!} \,$$

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
def J(n):
 sum1=0
 for m in range(0,70):
    sum1=sum1+(((-1)**m)*((x/2)**(2*m+n)))/((factorial(m))*factorial(m+n))
 return sum1
x=np.linspace(0,14,100)
for i in range(5):
    a=J(i)
    plt.plot(x,a)
plt.xlabel('$x$')
plt.ylabel('Bessel Function Values')
plt.title('Bessel Functions of the First Kind')
plt.legend(["$J0$","$J1$","$J2$","$J3$","$J4$"])
```
