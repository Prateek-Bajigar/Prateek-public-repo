```python
import matplotlib.pyplot as plt
import numpy as np
X=np.linspace(0,1,1000)
H=np.zeros([1000,1000])
h=X[2]-X[1]
for i in range(len(H)):
  try:
   H[i,i-1]=1
   H[i,i]=-2
   H[i,i+1]=1
  except:
    pass
H=(1/h**2)*H
H=((-1973**2)/(2*0.511*(10**6)))*H
eigenvalues,eigenvectors=np.linalg.eig(H)
idx=eigenvalues.argsort()[:]
eigenvalues=eigenvalues[idx]
eigenvectors=eigenvectors[:,idx]
A=np.sqrt(2)
for i in range(3):
  plt.figure(i)
  plt.subplot(3,2,1)
  plt.plot(X,(A*eigenvectors[:,i])*2,'-',color="blue")
  plt.ylabel(r'$\Psi(x)$')
  plt.xlabel("$x$")
  plt.title(f'PARTICLE IN A BOX n={i+1}')
  plt.subplot(3,2,2)
  plt.plot(X,(A*eigenvectors[:,i])**2,'-',color='cyan')
  plt.ylabel(r'$\Psi^2(x)$')
  plt.xlabel("$x$")
  plt.title("Probability Density")
plt.show()

print("Energy Eigenvalues :",eigenvalues[:3])
```
