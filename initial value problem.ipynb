# Question: Solving a Differential Equation Numerically

You are tasked with solving the following first-order ordinary differential equation (ODE) numerically:  

$$
\frac{dv}{dt} = 3v^2 - 5
$$

### Initial Condition:  
$$
v(0) = 0
$$

### Instructions:
1. Solve the differential equation using two numerical methods:  
   - **`odeint`** from `scipy.integrate`
   - **`solve_ivp`** from `scipy.integrate`
2. Use the following time domain for evaluation:  
   $
   t \in [-5, 5]
   $
3. Plot the solution obtained from both methods on the same graph:
   - Use a **blue line** for the `odeint` solution.
   - Use **red scatter points** for the `solve_ivp` solution.
4. Ensure the plot has gridlines for better readability.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp

def dvdt(t,v):
    return 3*v**2-5
v0=0

t=np.linspace(-5,5,1000)
sol_m1=odeint(dvdt,y0=v0,t=t,tfirst=True)
sol_m2=solve_ivp(dvdt,t_span=(-5,max(t)),y0=[v0],t_eval=t)
y1=sol_m1.T[0]
y2=sol_m2.y[0]
plt.plot(t,y1,'b')
plt.scatter(t,y2,color="red")
plt.grid(True)
```
