```python
import matplotlib.pyplot as plt
import numpy as np
X=np.linspace(-10,10,1000)
H=np.zeros([1000,1000])
B=10
h=X[2]-X[1]
k=100
for i in range(len(H)):
  try:
   H[i,i-1]=1
   H[i,i]=-2
   H[i,i+1]=1
  except:
    pass
#print(H)
H=(1/h**2)*H
H=((-197.3**2)/(2*940))*H
r=np.linspace(-10,10,1000)
V=np.zeros([1000,1000])
for i in range(1000):
  V[i,i]=(1/2)*k*r[i]**2
H=H+V;

eigenvalues,eigenvectors=np.linalg.eig(H)
idx=eigenvalues.argsort()[:]
eigenvalues=eigenvalues[idx]
eigenvectors=eigenvectors[:,idx]
eigenvalues[:3]

n=40
for i in range(n):
 if i==n-1:
  plt.plot(r,eigenvectors[:,i+1]**2,'b')
  plt.ylabel(r'$\Psi(x)$')
  plt.xlabel("$x$")
  plt.title("Wavefunction of Anharmonic oscillator")
  plt.grid(True)
  plt.ylim(-0.001,0.007)
  plt.show()

```
