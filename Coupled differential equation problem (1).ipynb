#### Find the solution of the following coupled differential equations:

$$\displaystyle
\frac{{dy_1}}{{dx}} = 5y_1 - 3y_2, \quad \frac{{dy_2}}{{dx}} = -6y_1 + 2y_2
$$

 $$ y_1(0)=2 ,  y_2(0)=1 $$



#### Solutions of this problem are
$$
y_1(x) = e^{-x} + e^{8x}, \quad y_2(x) = 2e^{-x} - e^{8x}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import solve_ivp
from scipy.integrate import odeint
```

```python
def dsdx(x,s):
    y1,y2=s
    return [5*y1-3*y2,-6*y1+2*y2]
y1_0=2
y2_0=1
s_0=(y1_0,y2_0)
x=np.linspace(0,5,1000)
sol_m2=solve_ivp(dsdx,y0=[y1_0,y2_0],method='DOP853'
,t_eval=x,rtol=10e-13,atol=10e-15,t_span=(0,max(x)))
y1=sol_m2.y[0]
y2=sol_m2.y[1]
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,'b')
plt.plot(x,y2,'r')
plt.legend(['y1','y2'])
plt.ylabel('$y(x)$',fontsize=10)
plt.xlabel('$x$',fontsize=10)
plt.title('Solutions using scipy',fontsize=10)
plt.grid(True)

x=np.linspace(0,5,1000)
y1=np.exp(-x)+np.exp(8*x)
y2=2*np.exp(-x)-np.exp(8*x)

plt.subplot(2,2,2)
plt.plot(x,y1,'b')
plt.ylabel('$y(x)$',fontsize=10)
plt.xlabel('$x$',fontsize=10)
plt.title('Original solutions',fontsize=10)
plt.plot(x,y2,'r')
plt.legend(['y1','y2'])
plt.tight_layout()
plt.grid(True)
```
