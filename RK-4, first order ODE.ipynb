### The RK-4 Algorithm: First order ODE

The Runge-Kutta 4th Order (RK-4) method solves the differential equation numerically by evaluating four slopes and computing a weighted average. The iterative process is described below:

1. **Step Size Calculation**:  
   The step size, $h$, defines how much to increment the value of $yn$ in each step. For each interval, the number of steps $n$ is given by:


   $$n = \frac{x_f - x_0}{h}$$


2. **Slope Calculations**:  
   In each iteration, we calculate four slopes:

   $$k_1 = h \cdot f(x_0, y_0)$$

   $$k_2 = h \cdot f\left(x_0 + \frac{h}{2}, y_0 + \frac{k_1}{2}\right)$$

   $$k_3 = h \cdot f\left(x_0 + \frac{h}{2}, y_0 + \frac{k_2}{2}\right)$$

   $$k_4 = h \cdot f(x_0 + h, y_0 + k_3)$$
   
   

3. **Updating the Solution**:  
   The new value of $y$ is computed as:

   $$y_{\text{new}} = y_0 + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$
   
   

4. **Iteration**:  
   After computing the new value of $y$, update $x_0$ and $y_0$ for the next iteration:

   $$x_0 = x_0 + h$$
   
   $$y_0 = y_{\text{new}}$$
   
   

5. **Continue Until Completion**:  
   The process continues until the interval from $x_0$ to $x_f$ is covered, and the final value of y is computed.
   
---

In summary, the RK-4 method computes a weighted average of four slopes at different points within an interval. It provides a more accurate solution to the differential equation compared to simpler methods like Euler's.


### Solving a problem in Python

- Consider the second order diffferential equation $$y'(x)=\frac{(x^2+y^2)}{(x+y)}$$ and given initial conditions

  $$y(0)=1$$ and we have to find $$y(2) $$


```python
# RK-4 METHOD TO SOLVE FIRST ORDER ODE

"""
Created on oct 8 20:35:27 2024
@author: prateek
"""

#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
def f(x,y):
    f=(x**2+y**2) /(x+y)  #THIS MEANS dy/dx=f=(whatever function is written here)
    return f
x0=eval(input("ENTER X0 : "))
xf=eval(input("ENTER XF : "))
y0=eval(input("ENTER Y0 : "))
h=eval(input("ENTER STEP SIZE (Less Than 1 and positive ) : "))
p=[]
q=[]

if xf<x0:
    h=-abs(h)

n=(xf-x0)/h

while n>0:
    k1=h*f(x0,y0)
    k2=h*f(x0+h/2,y0+k1/2)
    k3=h*f(x0+h/2,y0+k2/2)
    k4=h*f(x0+h,y0+k3)
    yf=y0+(1/6)*(k1+2*k2+2*k3+k4)
    
    p.append(x0)
    q.append(yf)
    
    y0=yf #so that on next iteration y0 takes value of yf which is in this iteration
    x0=x0+h
    n=n-1
print("VALUE OF THE GIVEN DIFF EQUATION AT xf is =",yf)
plt.plot(p,q,linestyle="solid",color='blue')
plt.title("RK-4 method for first order ODE",fontsize=20)
plt.xlabel(r"$x$",fontsize=18)
plt.ylabel(r"$y$",fontsize=18)
plt.legend([r"y(x)"],fontsize=10)
plt.grid(True)
plt.show()    
    
```
