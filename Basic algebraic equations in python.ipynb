## Linear equations

- Linear equations are the equations which represents straight lines. 
- They have only degree 1
- For example 

   - $4x-2=5$
   - $3x-1=10$
   - $4y-3x=10$


There are several methods for solving linear systems, including:
- Substitution method: Solve one equation for one variable and substitute it into the other equation.
- Elimination method: Add or subtract the equations to eliminate one variable.
- Matrix methods: Represent the system of equations using matrices and use matrix operations to find the solutions.

## Nonlinear  Equations

- A nonlinear equation does not form a straight line but generally forms a curve.
- They have degree of 2 or more.
- For example
   - $x^2+y^2=25$
   - $x^2+y=2$
- If they are not very complicated they can be used symbolically and even manually
- When they become quite tedious, methods like computer algebra system (CAS) approach and numerical methods like Newton-Paphson method,bisection methods are used to solve them.

## Solving algebraic equations in one variable

The solution to a linear system is the set of values that satisfies all the equations. Geometrically, the solution represents the intersection point(s) of the lines (in two dimensions) or the intersection of the planes (in three dimensions) defined by the equations.

Following are some equations in one variable
- $x-2-4=0$
- $x-3-6=0$
- $x^2-36=0$
- $x^3-3x-1=0$

Q1.**Solve** the equation $x-9=0$.

```python
from sympy import * # importing everything from sympy 
init_printing()  # For beautiful printing
x,y,z=symbols('x y z',real=True)  # Defining symbols
eq1=Eq(x-9,0)  # Defining equation x-9=0
sol=solve(eq1)  # using sympy's build-in function 'solve' to solve the equation 
sol 
```

Q2.**Solve** the equation $x^2-36=0$.

```python
eq2=Eq(x**2-36,0) # Defining equation x^2-36=0
sol=solve(eq2)
display(eq2) # Displaying equation 2 
display(sol)
```

# Algebraic equations in two variables 

Algebraic equations in two variables are typically represented in the form:

$$ax + by = c$$

"x" and "y" represent the variables in the equation. These variables can take on different values.
"a", "b", and "c" are coefficients or constants. "a" and "b" represent the coefficients of the variables "x" and "y", respectively, and "c" is a constant term.
<div class="alert alert-block alert-warning" style="color:brown;">
<b>Note:</b> To solve a system of equations in 'N' variables, we need N equations. For example, to solve this equation:
$$y+x=3$$ (having two variables x and y)

we need 2 equations. That is, we need one more equation involving only variables x and y, like the following:
$$y-x=6$$
</div>



```python
eq1 = Eq(y+x,3) #Define the equation y+x=3
eq2 = Eq(y -x , 6) #Define the equation y-x=6

solution = solve((eq1, eq2), (x, y))#Solving the system of equations
print("Equation 1 is ")
display(eq1)
print("Equation 2 is ")
display(eq2)
print("Values of x and y respectivelty are ")
display(solution)
```

#### Solving 3 equations with 3 variables

```python
eq1 = Eq(x**2 - y**2 ,14) # Define equation x^2-y^2=14
eq2 = Eq(2*x + 2*y , 10) # Defining the equation 2*x+2*y=10

solution = solve((eq1, eq2), (x, y))#Solving the system of equations
print("Equation 1 is ")
display(eq1)
print("Equation 2 is ")
display(eq2)
print("Values of x and y respectivelty are ")
display(solution)
```
