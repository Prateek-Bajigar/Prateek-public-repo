```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

A = float(input("Enter amplitude: "))
T = float(input("Enter time period (T): "))

x,n=symbols('x n')
T=T/2  #2T is the period of wave , so we need to give half value in T

N=eval(input("Enter Number of terms:"))
pi=np.pi
a0=(1/T)*(integrate((A/T)*x,(x,0,T)).evalf()+integrate((A/T)*(2*T-x),(x,T,2*T)).evalf())
an_values=[(A/(T**2))*(integrate(x*cos(n*pi*x/T),(x,0,T)).evalf()+integrate((2*T-x)*cos(n*pi*x/T),(x,T,2*T)).evalf()) for n in range(1,N+1)]
bn_values=[(A/(T**2))*(integrate(x*sin(n*pi*x/T),(x,0,T)).evalf()+integrate((2*T-x)*sin(n*pi*x/T),(x,T,2*T)).evalf()) for n in range(1,N+1)]
def f(value):
    sum=a0/2
    for i in range(1,N+1):
        sum+=an_values[i-1]*cos(i*pi*x/T)+bn_values[i-1]*sin(i*pi*x/T)
    return sum.subs(x,value)
a=np.linspace(0,10,100)
b=[f(i) for i in a]
plt.subplot(3,1,1)
plt.plot(a,b,'b')
plt.grid(True)
plt.title('Triangular wave by fourier series')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```
