
### Solve the first order ODE and find y at x=3.4.

$$\displaystyle \frac{dy}{dx}=x^2-y,y(0)=0$$


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib notebook
def f(y,x):
    return x**2-y
x=np.linspace(0,4,100)
y0=0
sol=odeint(f,y0,x)
plt.plot(x,sol[:,0],'r')
plt.ylabel("$y(x)$",fontsize=15)
plt.xlabel("$x$",fontsize=15)
plt.grid(True)
```
