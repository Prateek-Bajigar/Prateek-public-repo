### Solution-1

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
T=[300,400,500]
k=8.61*10e-5
x=np.linspace(0,5,1000)
for t in T:
    f_mb=1/(np.exp(x/(k*t)))
    plt.plot(x,f_mb)
plt.grid(True)
plt.legend(['300 K','400 K','500 K'])
plt.ylabel('$f_{Mb}(x)$')
plt.xlabel('Energy (eV)')
plt.title('Maxwell-Boltzmann Distribution')
```

### Solution-2 a)

```python
#%matplotlib notebook
T=[0,10,50,500]
mu=0.025
x=np.linspace(0,0.5,1000)
for t in T:
 f_fd=1/(np.exp((x-mu)/(k*t))+1)
 plt.plot(x,f_fd)
plt.grid(True)
plt.legend(['0 K','10 K','50 K','500 K '])
plt.ylabel('$f_{Fd}(x)$')
plt.xlabel('Energy (eV)')
plt.title('Fermi-Dirac Distribution')
```

### Solution-3 a)

```python
T=[300,400,500]
mu=-0.15
x=np.linspace(-0.14,0.15,1000) # don't take x=-0.15
for t in T:
 f_Be=1/(np.exp((x-mu)/(k*t))-1)
 plt.plot(x,f_Be)
plt.grid(True)
plt.legend(['300 K','400 K','500 K '])
plt.ylabel('$f_{Be}(x)$')
plt.xlabel('Energy (eV)')
plt.title('Bose-Einstein Distribution')
```
