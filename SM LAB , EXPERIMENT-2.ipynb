### 1.Plot Cv/3R vs T using
a) Dulong-Petit Law

The Dulong-Petit Law of Gases is given by the equation:

$$\displaystyle
\frac{C_v}{3R} = 1
$$

b) Einstein Law,$T_e=100K$

The Einstein Law of Gases is given by the equation:

$$\displaystyle
\frac{C_v}{3R} = \left(\frac{T_e}{T}\right)^2 \frac{e^{\frac{T_e}{T}}}{\left(e^{\frac{T_e}{T}} - 1\right)^2}
$$

Where:
- $ C_v$  is the molar specific heat capacity at constant volume,
- $ R $ is the ideal gas constant,
- $ T $ is the temperature,
- $ T_e $ is the Einstein temperature.



c) Debye specific heat model , $T_d=100K$

The debye Law of Gases is given by the equation:

$$\displaystyle
\frac{C_v}{3R} = 3\left(\frac{T_d}{T}\right)^3 \ \int_{0}^{\frac{T_d}{T}} \frac{x^4  e^x}{\left(e^x - 1\right)^2} \, dx
$$

where:
- $T_d$  is the Debye temperature.




```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
import warnings
warnings.filterwarnings("ignore") # supresses all warnings
%matplotlib notebook
t=np.linspace(1,350,1000) #starting with 0 gives error
te=100
td=100

y1=np.ones(1000) # to create array of 1000 1's
y2=(te/t)**2*(np.exp(te/t)/(np.exp(te/t)-1)**2)
def g(x):
    return (x**4*np.exp(x))/((np.exp(x)-1)**2)

# Integrate the function from intial to final range to get area
integrals=[]
rest_part=[]
for i in t:
    integral, err = quad(g, 0, td / i)
    integrals.append(integral)
for i in t:
    w=3*(i/td)**3
    rest_part.append(w)
y3=[integrals[i]*rest_part[i] for i in range(len(integrals))]
plt.plot(t,y1)
plt.plot(t,y2)
plt.plot(t,y3)
plt.xlabel('T(K)',fontsize=13)
plt.ylabel(r"$\frac{C_v}{3R}$",fontsize=20)
plt.legend(['Dulong-Petit law',"Einstein's law","Debye's law"])
plt.grid(True)
```

### 2. Plot Debye specific heat model for given three materials.


|Material | Debye's temperature(K) |
|-----------------|-------------|
| Si          | 645     |
| Cu           |343      |
| Pb           | 105      |


```python
t=np.linspace(1,800,1000)
T=[645,343,105]

def g(x):
    return (x**4*np.exp(x))/((np.exp(x)-1)**2)

# Integrate the function from intial to final range to get area
for t_ in T:
 integrals=[] # array for integer values for each t value
 rest_part=[] # array for rest of the part for each t value
    
 for i in t:
     integral, err = quad(g, 0, t_ / i)
     integrals.append(integral)
   
 for i in t:
     w=3*(i/t_)**3
     rest_part.append(w)
    
 y3=[integrals[i]*rest_part[i] for i in range(len(integrals))]
 plt.plot(t,y3)
 plt.grid(True)
plt.xlabel('T(K)',fontsize=13)
plt.title("Debye's model for Si,Cu and Pb" )
plt.ylabel(r"$\frac{C_v}{3R}$",fontsize=20)
plt.legend(['Si-645 K','Cu-343 K','Pb-105 K'])
```
