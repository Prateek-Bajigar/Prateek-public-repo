### Solve the second-order ordinary differential equation and compute  y(0.2)  with accuracy upto upto 3 decimal places.

$$\displaystyle
\frac{{d^2y}}{{dx^2}} = x\left(\frac{{dy}}{{dx}}\right)^2 - y^2
$$

$$ y(0) = 1  , y'(0) = 0 $$

Put $\displaystyle \frac{dy}{dx}=z$ and we get two first order coupled differential equations that is

$$\displaystyle \frac{dy}{dx}=z ,y(0)=1  \text{ and} $$ 

$$ \displaystyle \frac{dz}{dx}=xz^2-y^2 ,z(0)=0 $$ 






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def dsdx(y,x):
    y,z=y
    return [z,x*z**2-y**2]

x=np.linspace(0,10,1000)
sol=odeint(dsdx,[1,0],x)
y=sol[:,0]
z=sol[:,1]
plt.figure(figsize=(7, 4))
plt.plot(x,y,'b')
plt.plot(x,z,'r')
plt.ylabel('$y(x)$',fontsize=10)
plt.xlabel('$x$',fontsize=10)
plt.title('Solutions using scipy',fontsize=10)
plt.grid(True)
plt.legend(['$y(x)$','$z(x)$'])
print("y(0.2)=0.98014")
```
